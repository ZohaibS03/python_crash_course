```python
def city_country(city, country):
    """Function to return city and country formatted nicely"""
    return f'{city}, {country}'.title()

city_country('Santiago', 'Chile')
```

```python
def make_album(artist_name, album_title, songs=None):
    """Function to create a dictionary about an album"""
    album = {'artish_name': artist_name, 
             'album_title': album_title,
             }
    if songs:
        album['songss'] = songs
    return album

print("Enter q at any time to quit")
while True:
    art_name = input('Artist name: ')
    if art_name == 'q':
        break
    alb_title = input('Album title: ')
    if alb_title == 'q':
        break
    songs = input('Number of songs:') 
    if songs.isnumeric(): # If number is given for songs then all args passed
        print(make_album(art_name, alb_title, songs))
        continue 
    elif songs == 'q': # If q then quits
        break
    else:
        print(make_album(art_name, alb_title)) # Only artist and album passed
         

```
