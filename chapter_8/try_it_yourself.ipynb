```python
def shirt_summary(shirt_size, shirt_message):
    """Function to ouput passed arguments"""
    print(f'You have a size {shirt_size} shirt, and the message on it is '
            f'{shirt_message}') # Multiline for readability; no space issues

shirt_summary('large', 'peacemaker!')
shirt_summary(shirt_message='i love cats!', shirt_size='x-small')
```

```python
def shirt_summary(shirt_size='large', shirt_message='I love Python!'):
    """Function to ouput passed arguments"""
    print(f'You have a size {shirt_size} shirt, and the message on it is '
            f'{shirt_message}')

shirt_summary()
shirt_summary(shirt_size='medium')
shirt_summary(shirt_size='small', shirt_message='I love Dogs!')
```

```python
def describe_city(city, country='UK'):
    """A function to print information about a City and where it is"""
    print(f'{city} is in {country}.')

describe_city('London')
describe_city('Manchester')
describe_city('Paris', 'France')
```
